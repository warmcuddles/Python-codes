```python
def func(text):

    li1=text.split('\n')

    li = [x for x in li1 if x]

    li2 = [x.replace(' \uf00d', '') for x in li]
    li3 = [x.replace(' \uf00c', '') for x in li2]
    #print(li3)
    chunks = [li3[x:x+5] for x in range(0, len(li3), 5)]
    
    df = pd.DataFrame.from_records(chunks)
    df.columns = ['Questions', 'A', 'B', 'C', 'D']
    
    ans = text.split('\n')
    ans = [x for x in ans if x]
    #print(ans)
    answer = []
    for i in ans:
        if '\uf00c' in i:
            if i[:1]=='1':
                answer.append('A')
            elif i[:1]=='2':
                answer.append('B')
            elif i[:1]=='3':
                answer.append('C')
            if i[:1]=='4':
                answer.append('D')
    df['A'] = df['A'].str[3:]
    df['B']=df['B'].str[3:]
    df['C'] = df['C'].str[3:]
    df['D']=df['D'].str[3:]   
    
    df['Correct_Choice'] = pd.Series(answer, index=df.index)
    df['Question_no'] = df['Questions'].str[:3]
    df = df[['Question_no', 'Questions', 'A', 'B', 'C', 'D', 'Correct_Choice']]
    df['Questions'] = df['Questions'].str[4:]
    return(df)
df2 = func(text4)
```
