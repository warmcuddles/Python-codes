A point at angle theta on the circle whose centre is $(x0,y0)$ and whose radius is r is $(x0 + r*cos(theta),y0 + r*sin(theta)$. Now choose theta values evenly spaced between 0 and 2pi.



```python
import math
import numpy as np
def circ(c,r):
    x0 = c[0]
    y0 = c[1]
    li = []
    distlist = np.linspace(0, 2*math.pi, num=50)
    for theta in distlist:
        t = (x0 + r*math.cos(theta), y0 + r*math.sin(theta))        
        li.append(t)
    
    import matplotlib.pyplot as plt
    plt.rcParams['figure.figsize']=(8.0, 8.0)
    circle = plt.Circle((10, 10), 5, color='blue')
    fig, ax = plt.subplots()
    ax.add_artist(circle)
    # change default range so that new circles will work
    ax.set_xlim((0, 20))
    ax.set_ylim((0, 20))
    
    for i in range(len(li)):
        x=li[i][0]
        y=li[i][1]
        plt.plot(x,y, 'o', color='black')
    
    
    print(li)

    
circ((10, 10), 5)
```
